```
def q5():    
  a = 14.2-11.6
  b = 13.5-15.9
  c = 15.9-18.8
  d = 17.7-20.4
  e = 18.6-17.7
  dados = {'a': a, 'b':b, 'c':c, 'd':d, 'e':e}
  print(dados)
  for alt, resp in dados.items():
    if resp == min(dados.values()):
      return str(f'Alternativa {alt.upper()}, {resp}')
   

q5()
```
